```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

excel_path1='jfreechart-metric5&6.xls'
excel_path2='metric 5&6 docker.xlsx'
excel_path3='metric 5&6 collection.xlsx'
df1 = pd.read_excel(excel_path1,na_values='0').head(2)
df1 = df1[['DLOC','BMI']]
df2 = pd.read_excel(excel_path2,na_values='0')
df2 = df2[['DLOC','BMI']]
df3 = pd.read_excel(excel_path3,na_values='0')
df3 = df3[['DLOC','BMI']]
df = pd.concat([df1,df2,df3])
corr=df.corr(method='pearson',min_periods=1)
print(corr)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

excel_path1='docker1.2.2.xlsx'
df = pd.read_excel(excel_path1,na_values='0')
df = df[['Mutation Cov','Statement Cov.','Branch Cov.','Cxty']]
corr = df.corr(method='pearson',min_periods=1)
df.plot(x='Statement Cov.', y='Mutation Cov', kind='scatter')
plt.savefig('docker_1_3.jpg')
df.plot(x='Branch Cov.', y='Mutation Cov', kind='scatter')
plt.savefig('docker_2_3.jpg')
df.plot(x='Statement Cov.', y='Cxty', kind='scatter')
plt.savefig('docker_1_4.jpg')
df.plot(x='Branch Cov.', y='Cxty', kind='scatter')
plt.savefig('docker_2_4.jpg')
sns.pairplot(df[['Mutation Cov','Statement Cov.','Branch Cov.']])
plt.savefig('docker_123.jpg')
sns.pairplot(df[['Cxty','Statement Cov.','Branch Cov.']])
plt.savefig('docker_124.jpg')
plt.show()
plt.close()
```
